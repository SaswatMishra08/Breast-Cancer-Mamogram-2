```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read input image
img = cv2.imread(r"C:\Users\Siddharth\Desktop\Mammography_DataSet\Classified\Cancer_Severity\Malignant\mdb028.png")

# Resizing and cropping
resized_img = cv2.resize(img, (256, 256))
cropped_img = resized_img[50:200, 50:200]

# Normalization and rescaling
normalized_img = cv2.normalize(cropped_img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

# Data augmentation
flipped_img = cv2.flip(normalized_img, 1)  # 1 for horizontal flip

# Noise reduction
blur_img = cv2.GaussianBlur(flipped_img, (3, 3), 0)

# Color correction
corrected_img = cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB)

# Show original and processed images
plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.subplot(2, 3, 3), plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.subplot(2, 3, 4), plt.imshow(normalized_img)
plt.title('Normalized Image')
plt.subplot(2, 3, 5), plt.imshow(flipped_img)
plt.title('Flipped Image')
plt.subplot(2, 3, 6), plt.imshow(corrected_img)
plt.title('Corrected Image')

plt.show()
```
